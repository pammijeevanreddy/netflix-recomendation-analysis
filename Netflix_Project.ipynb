```python
df['count_countrys']=df['country'].str.split(',').str.len()
df['cast_count']=df['cast'].str.split(',').str.len()
df['director_count']=df['director'].str.split(',').str.len()
```

```python
df.head()
```

```python
duration={
    'Kids' : ['TV-Y','TV-Y7','TV-G','G','TV-Y7-FV'],
    'Teen' : ['TV-PG','PG','PG-13','TV-14'],
    'Adult' : ['TV-MA','R','NC-17','NR','UR']
}

def jeevan(x):
    for i,j in duration.items():
        if x['rating'] in j:
            return i

df['category']=df.apply(lambda x: jeevan(x), axis=1)
```

```python
col='country'
```

```python
df[col]=df[col].astype(str).str.lower()
```

```python
df[col]=df[col].str.split(',')
```

```python
df[col] = df[col].apply(lambda x: [i.strip() for i in x if i.strip() != ''])
```

```python
# Explode into seperate rows
df=df.explode(col)
```

```python
df[col]=df[col].str.title()
```

```python
df['country'].unique()
```

```python
df.reset_index(inplace=True)
df
```

```python
pivot_table = pd.pivot_table(
    df,
    index='release_year',
    columns='country',
    values='show_id',
    aggfunc='count',
    fill_value=0
)
pivot_table['Grand Total'] = pivot_table.sum(axis=1)
pivot_table.loc['Grand Total'] = pivot_table.sum(axis=0)
# pivot_table = pivot_table.T
pivot_table
```

```python
pivot = pivot_table.sort_values(ascending = False, by = "Grand Total").head(16)
pivot
```

```python
T_pivot = pivot.T
T_pivot
```

```python
T_pivot = T_pivot.sort_values(ascending = False, by = "Grand Total").head(6)
T_pivot
```

```python
T_pivot.drop("Grand Total", axis = 1, inplace = True)
T_pivot.drop("Grand Total", axis = 0, inplace = True)
```

```python
pivot = T_pivot.T
```

```python
# pivot = pivot.T
pivot
```

```python
pivot_index=pivot.reset_index()
pivot_index
```

```python
pivot_melted = pivot_index.melt(id_vars='release_year', var_name='country', value_name='count')
```

```python
sns.lineplot(data=pivot_melted, x='release_year', y='count', hue='country', marker='o')
plt.title("Trend Of Netflix Titles by country Over The Year")
plt.show()
```
